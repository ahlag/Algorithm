## Personal Solution
### Runs through the list twice, the first time to check the length and the second time to output the list starting from the middle

```python
class Solution:
    def middleNode(self, head):
        
        node = head
        length = 0
        while node:
            node = node.next
            length += 1
            
        node = head
        
        if length == 1:
            return node
        
        middle = length // 2
        cur_index = 0
        while node:
            node = node.next
            cur_index += 1
            if cur_index == middle:
                return node
```

```python
class Solution(object):
    def middleNode(self, head):
        A = [head]
        while A[-1].next:
            A.append(A[-1].next)
        return A[len(A) / 2]
```

```python
class Solution(object):
    def middleNode(self, head):
        slow = fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow
```

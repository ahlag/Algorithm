```python
class Solution:
    def mergeTwoLists(self, l1, l2):
        # maintain an unchanging reference to node ahead of the return node.
        prevhead = ListNode()
        
        head = prevhead
        
        while l1 and l2:
            
            if l1.val >= l2.val:
                head.next = l2
                l2 = l2.next
            else:
                head.next = l1
                l1 = l1.next
            head = head.next
            
        head.next = l1 if l1 is not None else l2
        
        return prevhead.next
```

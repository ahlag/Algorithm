```python
class Solution:
    def transpose(self, A: List[List[int]]) -> List[List[int]]:
        
        n, m = len(A), len(A[0])
        B = [[0] * n for _ in range(m)]
        
        for i in range(m):
            for j in range(n):
                if i != j:
                    B[i][j] = A[j][i]
                else:
                    B[i][j] = A[i][j]
                
        return B
```

```python
class Solution:
    def transpose(self, A: List[List[int]]) -> List[List[int]]:
        
        res = []
        
        for i in range(len(A[0])):
            tmp = []
            for j in range(len(A)):
                tmp.append(A[j][i])
            res.append(tmp)
                
        return res
```

### C++

```python
class Solution {
public:
    vector<vector<int>> transpose(vector<vector<int>>& A) {
        int N = A.size(), M = A[0].size();
        vector<vector<int>> B(M, vector<int>(N, 0));
        
        for(int i = 0; i < M; i++) {
            for(int j = 0; j < N; j++) {
                B[i][j] = A[j][i];
            }
        }
        return B;
    }
};
```

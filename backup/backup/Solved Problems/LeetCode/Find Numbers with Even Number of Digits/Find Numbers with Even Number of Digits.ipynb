```python
class Solution:
    def findNumbers(self, nums) -> int:
        
        ec = 0
        for num in nums:
            cnt = 0
            while num:
                num //= 10
                cnt += 1
            if cnt % 2 == 0:
                ec += 1
                
        return ec
```

## Personal Solution

```python
keyboard = "abcdefghijklmnopqrstuvwxyz"
word = "cba"

def calculateTime(keyboard, word):
        """
        :type keyboard: str
        :type word: str
        :rtype: int
        """
        index = [i for i in range(26)]
        typeset = dict(zip(keyboard, index))

        i = 0
        dis = 0
        prev_char = 0

        for char in word:
            dis += abs(typeset[char] - prev_char)
            prev_char = typeset[char]
        
        return dis
    
calculateTime(keyboard, word)
```

## Clean Solution

```python
keyboard = "abcdefghijklmnopqrstuvwxyz"
word = "cba"

def calculateTime(keyboard, word):
        """
        :type keyboard: str
        :type word: str
        :rtype: int
        """
        dic = {}
        distance = prev = 0
        
        for idx, ch in enumerate(keyboard):
            dic[ch] = idx
            
        for ch in word:
            distance += abs(prev - dic[ch])
            prev = dic[ch]
        
        return distance
    
calculateTime(keyboard, word)
```

## Cleaner Solution

ini (keyboard[0]) -> b -> a
Creating a pair \
a   c   b \
   \
c   b   a

```python
keyboard = "abcdefghijklmnopqrstuvwxyz"
word = "cba"

def calculateTime(keyboard, word):
    kbd = {c:i for i, c in enumerate(keyboard)}
    return sum(abs(kbd[c1]-kbd[c2]) for c1,c2 in zip(keyboard[0]+word[:-1], word))
    
calculateTime(keyboard, word)
```

## Personal Solution

```python
class Solution:
    def isArmstrong(self, N: int) -> bool:
        
        product = N
        exponent = 0
        while product:
            product = product // 10
            exponent += 1
            
        product = N
        res = 0
        while product:
            digit = product % 10
            product = product // 10
            res += digit ** exponent
        
        return res == N
```

## Using String

```python
class Solution:
    def isArmstrong(self, N: int) -> bool:
        return sum(map(lambda x: int(x)**len(str(N)), str(N))) == N
```

## Not Using String

```python
class Solution:
    def isArmstrong(self, N: int) -> bool:
        digits = []
        n = N
        while n:
            n, digit = divmod(n, 10)
            digits.append(digit)

        return sum(map(lambda x: x**len(digits), digits)) == N
```

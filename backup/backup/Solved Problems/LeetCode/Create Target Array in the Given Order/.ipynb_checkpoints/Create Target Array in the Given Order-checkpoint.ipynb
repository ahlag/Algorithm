```python
def createTargetArray(nums, index):
        """
        :type nums: List[int]
        :type index: List[int]
        :rtype: List[int]
        """
        size = len(nums)
        target = [None] * size
        for i in range(size):
            ind = index[i]
            if target[ind] is None:
                target[ind] = nums[i]
            else:
                j = size - 1
                while j > ind:
                    target[j], target[j-1] = target[j-1], target[j]
                    j -= 1
                target[ind] = nums[i]
                
        return target
```

```python
nums = [0,1,2,3,4]
index = [0,1,2,2,1]
createTargetArray(nums,index)
```

```python
def createTargetArray(nums, index):
    target =[]
    for i in range(len(nums)):
        if index[i] == len(target) :
            target.append(nums[i])
        else:
            target = target[:index[i]] + [nums[i]] + target[index[i]:]
    return target
```

```python
createTargetArray(nums,index)
```

## Personal Solution

```python
class OrderedStream:

    def __init__(self, n: int):
        self.slots = [None] * n
        self.max_size = n
        self.ptr = 0

    def insert(self, id: int, value: str):
        
        id = id - 1
        self.slots[id] = value
        
        output = []
        for i in range(self.ptr, self.max_size):
            if self.slots[i] is None:
                break
            output.append(self.slots[i])
            self.ptr += 1
        return output
```

## Clean Solution

```python
class OrderedStream:

    def __init__(self, n: int):
        self.data = [None]*n
        self.ptr = 0 # 0-indexed 

    def insert(self, id: int, value: str):
        id -= 1 # 0-indexed 
        self.data[id] = value 
        if id > self.ptr:
            return [] # not reaching ptr 
        
        while self.ptr < len(self.data) and self.data[self.ptr]: 
            self.ptr += 1 # update self.ptr 
            
        return self.data[id:self.ptr]
```

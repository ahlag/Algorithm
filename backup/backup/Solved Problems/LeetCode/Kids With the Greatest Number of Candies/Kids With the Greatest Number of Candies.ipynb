## Personal Solution

```python
class Solution:
    def kidsWithCandies(self, candies, extraCandies: int):
        
        mc = max(candies)
        n = len(candies)
        output = [False] * n
        
        for i in range(n):
            if candies[i] + extraCandies >= mc:
                output[i] = True
                
        return output
```

## Clean Solution

```python
def kidsWithCandies(self, candies, extraCandies: int):
    maxCandies = max(candies)
    return [candy+extraCandies >= maxCandies for candy in candies]
```

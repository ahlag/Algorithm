```python
def findJudge(N, trust):
    
    if len(trust) < N - 1:
        return -1
    
    outdegree = [0] * (N + 1)
    indegree = [0] * (N + 1)
    
    for person in trust:
        a, b = person
        outdegree[a] += 1
        indegree[b] += 1
        
    for i in range(1, N+1):
        if outdegree[i] == 0 and indegree[i] == N-1:
            return i
            
    return -1
```

```python
def findJudge(N, trust):
    person = [0] * (N+1)
        
    for edge in trust:
        a, b = edge
        person[a] -= 1
        person[b] += 1 
        
    for i in range(1, N+1):
        if person[i] == N - 1:
            return i
            
    return -1
```

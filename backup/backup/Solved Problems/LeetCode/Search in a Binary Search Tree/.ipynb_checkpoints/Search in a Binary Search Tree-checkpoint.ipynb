```python
class Solution:
    def searchBST(self, root, val: int):
        
        if root is None:
            return None
        
        stack = [root]
        
        while stack:
            node = stack.pop()
            if node.val == val:
                return node
            if node.right is not None:
                stack.append(node.right)
            if node.left is not None:
                stack.append(node.left)
```

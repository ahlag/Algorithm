## Personal Solution

```python
class Solution:
    def diagonalSum(self, mat) -> int:
        
        total_sum = 0
        n = len(mat)-1
        
        for i in range(len(mat)):
            if i == n-i:
                total_sum += mat[i][i]
            else:
                total_sum += mat[i][i] + mat[i][n-i]
                
        return total_sum
```

```python
# Python
class Solution:
    def diagonalSum(self, A):
        n   = len(A)
        res = 0
        for i in range(n):
            res += A[i][i] + A[i][n-1-i]
        # x: Index of Element Repeated (when n is Odd)
        x = (n-1)//2
        return res - A[x][x] if n&1 else res
```

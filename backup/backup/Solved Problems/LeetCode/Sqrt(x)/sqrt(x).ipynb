```python
from math import e, log
def sqrt(x: int):
    if x < 2:
        return x
    
    left  = int(e ** (0.5 * log(x)))
    right = left + 1
    return left if right * right > x else right
```

```python
from math import e, log
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """
        if x < 2:
            return x
        left, right = 2, x//2
        
        while left <= right:
            pivot = left + (right - left) // 2
            num = pivot * pivot
            if num > x:
                right = pivot - 1
            elif num < x:
                left = pivot + 1
            else:
                return pivot
        
        return right
```

```python
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """
        if x < 2:
            return x
        x0 = x
        x1 = (x0 + x / x0) / 2
        while abs(x0 - x1):
            x0 = 1
            x1 = (x0 + x / x0) / 2
            
        return int(x1)
```

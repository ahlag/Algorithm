```python
class Solution:
    def titleToNumber(self, s: str) -> int:
        alphabets = { chr(64 + i) : i for i in range(1,27)}
        res = alphabets[s[-1]] 
        cnt = 1
        for i in range(len(s)-1, 0, -1):
            res += alphabets[s[i-1]] * (26 ** cnt)
            cnt += 1
            
        return res
```

```python
class Solution:
    def oddCells(self, n: int, m: int, indices) -> int:
        
        matrix = [[0] * m for i in range(n)]
        
        for row, column in indices:
            
            for i in range(0,m):
                matrix[row][i] += 1
                
            for i in range(0,n):
                matrix[i][column] += 1
           
        cnt = 0
        for i in range(n):
            for j in range(m):
                if matrix[i][j] % 2 != 0:
                    cnt += 1
                    
        return cnt
```

## Numpy

```python
import numpy as np


def oddCells(n, m, indices):
    matrix = np.zeros(shape=(n, m))
    for pair in indices:
        row_idx, col_idx = pair
        matrix[row_idx] = [x + 1 for x in matrix[row_idx]]
        for row in matrix:
            row[col_idx] += 1
            
    matrix_flat = matrix.flatten()
    
    return len(list(filter(lambda x: x % 2 == 1, matrix_flat)))
```

## Two Lists

1. Initialize first list of zeros length n and second list of zeros length m.
2. Iterate over indexes.
3. increment value at index[0] in first list by 1 and increment value at index[1] in second list by 1.
4. Sum list comprehension nesting iteration of values in second list within iteration of values in first list appending 1 for every (i+j) % 2.

https://leetcode.com/problems/cells-with-odd-values-in-a-matrix/discuss/751248/O(m%2Bn)-Time-and-Space-solution

```python
class Solution:
    def oddCells(self, n: int, m: int, indices) -> int:
        x, y = [0] * n, [0] * m                             # two lists - one for rows, one for columns
        for i in indices:                                   # for each index...
            x[i[0]] += 1                                    # add to it's row
            y[i[1]] += 1                                    # add to it's column
        return sum([1 for j in y for i in x if (j+i) % 2])  # tally matrix
```

```python
class Solution:
    def oddCells(self, n: int, m: int, indices) -> int:
        r = [0]*n # total rows filled up with zeros
        c = [0]*m # total cols filled up with zeros
        for i, j in indices: # to account for indices
            r[i]+=1
            c[j]+=1
        er = 0 # even numbers in r
        for i in r:
            if i%2==0:
                er+=1
        ec = 0 # even numbers in c
        for j in c:
            if j%2==0:
                ec+=1
        odr = n-er # odd numbers is r
        odc = m-ec # odd numbers in c
        return er*odc + odr*ec # odd = even + odd # combinations
```

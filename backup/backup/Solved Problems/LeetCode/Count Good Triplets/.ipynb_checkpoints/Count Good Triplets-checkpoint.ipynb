## Brute Force

```python
class Solution:
    def countGoodTriplets(self, arr, a: int, b: int, c: int) -> int:
        
        cnt = 0
        for i in range(0, len(arr)-2):
            for j in range(i+1, len(arr)-1):
                for k in range(j+1, len(arr)):
                    
                    if (abs(arr[i] - arr[j]) <= a) \
                        and (abs(arr[j] - arr[k]) <= b) \
                        and (abs(arr[i] - arr[k]) <= c):
                            cnt += 1
                
        return cnt
```

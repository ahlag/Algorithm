```python
def removeVowels(S: str) -> str:
    # Think about your edge cases
    if len(S) == 0:
        return ""
        
    # Consider all the data structures to store, here string seems the easiest
    result_string = ""
    # List gave Runtime: 28 ms, faster than 59.87% of Python3
    #vowels = ['a','e','i','o','u']
    # Set gave: Runtime: 20 ms, faster than 96.69% of Python3 
    vowels = ('a','e','i','o','u')
    # Again best to try on your machine, since Leet code performance can vary
    
    for char in S:
        if char not in vowels:
            result_string += char
```

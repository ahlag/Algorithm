## Personal Solution

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        
        array = [0] * n
        res = 0
        res = start
        for i in range(1, n):
            array[i] = start + 2 * i
            res = res ^ array[i]
            
        return res
```

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        ans=0
        for i in range(n):
            ans^=start+(2*i)
        return ans
```

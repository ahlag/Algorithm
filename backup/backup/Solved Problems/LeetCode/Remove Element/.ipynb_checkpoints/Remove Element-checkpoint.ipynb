```python
def removeElement(nums, val):
    j = 0
    for i in range(len(nums)):
        if nums[i] != val:
            nums[j] = nums[i]
            j += 1
    return j
```

```python
def removeElement(nums, val):
    
    i = 0
    n = len(nums) 
    while (i < n):
        if nums[i] == val:
            nums[i] = nums[n-1]
            n -= 1
        else:
            i += 1
            
    return n 
```

```python
class Solution:
    def restoreString(self, s: str, indices) -> str:
        
        arr = [''] * len(s)
        j = 0
        for i in indices:
            arr[i] = s[j]
            j += 1
            
        return "".join(arr)
```

```python
class Solution:
    def restoreString(self, s: str, indices) -> str:
        ans = ""
        for i in range(len(indices)):
            ans += s[indices.index(i)]
        return ans
```

## Selection Sort

```python
class Solution:
    def restoreString(self, s: str, indices) -> str:
        s1 = list(s)
        for i in range(0, len(indices)):
            #assume the first element is the lowest one in the unsort part
            lowest_idx = i
            # iterate the unsorted part except first one
            for j in range(i+1, len(indices)):
                #if we find the lowest one in the unsorted part of indices
                #record the real index 
                if indices[j] < indices[lowest_idx]:
                    lowest_idx = j
            indices[i], indices[lowest_idx] = indices[lowest_idx], indices[i]
            s1[i], s1[lowest_idx] = s1[lowest_idx], s1[i]
        return ''.join(s1)
```

## Brute Force

```python
class Solution:
    def numIdenticalPairs(self, nums) -> int:
        
        cnt = 0
        for i in range(len(nums)-1, -1, -1):
            for j in range(i-1, -1, -1):
                if nums[i] == nums[j] and i > j:
                    cnt += 1
                    
        return cnt
```

```python
class Solution:
    def numIdenticalPairs(self, nums) -> int:
        if len(nums) == len(set(nums)):
            return 0
        frequency = {}
        for index in nums:
            if index in frequency.keys():
                frequency[index] += 1
            else:
                frequency[index] = 1
        answer = 0
        for key, val in frequency.items():
            answer += (val*(val - 1)) // 2
        return answer
```

```python
class Solution(object):
    def uniqueMorseRepresentations(words):
        """
        :type words: List[str]
        :rtype: int
        """
        morse_code = [".-","-...","-.-.","-..",".","..-.","--.","....","..",".---","-.-",".-..","--","-.","---",".--.","--.-",".-.","...","-","..-","...-",".--","-..-","-.--","--.."]
        alphabet = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z"]
        morse_dict = dict(zip(alphabet,morse_code))
        
        codes = []

        for word in words:
            decipher = ""
            for i in word:
                decipher += morse_dict[i]
            codes.append(decipher)
        
        return len(set(codes))
```

```python
words = ["gin", "zen", "gig", "msg"]

class Solution(object):
    def uniqueMorseRepresentations(words):
        MORSE = [".-","-...","-.-.","-..",".","..-.","--.",
                 "....","..",".---","-.-",".-..","--","-.",
                 "---",".--.","--.-",".-.","...","-","..-",
                 "...-",".--","-..-","-.--","--.."]

        seen = {"".join(MORSE[ord(c) - ord('a')] for c in word)
                for word in words}

        return len(seen)
    
Solution.uniqueMorseRepresentations(words)
```

```python
def uniqueMorseRepresentations(self, words: 'List[str]') -> 'int':
        mourse_codes = [".-","-...","-.-.","-..",".","..-.","--.","....","..",".---","-.-",".-..","--","-.","---",".--.","--.-",".-.","...","-","..-","...-",".--","-..-","-.--","--.."]
        
        alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
        
        dct = {i:j for i,j in zip(alphabet, mourse_codes)}
    
        words_morse = []
        for word in words:
            words_morse.append(''.join(list(map(dct.get, list(word)))))
        
        return len(set(words_morse))
```

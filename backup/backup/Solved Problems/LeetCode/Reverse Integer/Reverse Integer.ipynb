```python
# Personal Solution
class Solution:
    def reverse(self, x: int) -> int:
        
        limit = 2**31
        upper_bound = limit - 1
        lower_bound = -limit
        
        if x < 0:
            div = (-1) * x
        else:
            div = x

        rem = 0
        
        while div > 0:
            num = div
            div = num // 10
            rem = num % 10 + rem * 10
            if rem < lower_bound or rem > upper_bound:
                return 0
        
        return (-1)*rem if x < 0 else rem
```

```python
# Clean Solution
class Solution:
    def reverse(self, x: int) -> int:
        new_val = 0
        sign = 1 if x > 0 else -1
        x = abs(x)
        
        while x:
            n, rem = divmod(x, 10)
            new_val = new_val * 10 + rem
            x = n
            
        return sign * new_val if new_val < 2 ** 31 else 0
```

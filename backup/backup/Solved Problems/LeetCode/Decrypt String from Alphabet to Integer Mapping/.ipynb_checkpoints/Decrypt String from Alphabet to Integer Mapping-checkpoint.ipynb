## Personal Solution

```python
class Solution:
    def freqAlphabets(self, s: str) -> str:
        
        mapping = {str(i - 96): chr(i) for i in range(97,106)}
        
        for i in range(106,123):
            mapping[str(i-96) + '#'] = chr(i)
        
        n = len(s)
        isHashtag = [False] * n
        for i in range(n):
            if s[i] == '#':
                isHashtag[i-2] = True
        
        output = ''
        i = 0
        while i < n:
            if isHashtag[i] is True:
                output += mapping[s[i:i+3]]
                i += 3
            else:
                output += mapping[s[i]]
                i += 1
                
        return output
```

## Replace Solution

```python
class Solution:
    def freqAlphabets(self, s: str) -> str:
        for i in range(10,27):
            s=s.replace(str(i)+'#',chr(96+i))
        for i in range(1,10):
            s=s.replace(str(i),chr(96+i))
        return s
```

## Backward to Fornt

```python
class Solution:
    def freqAlphabets(self, s: str) -> str:
        new_arr = []
        i = len(s)-1
        while i >= 0:
            if s[i] == "#":
                new_arr.insert(0, chr(int(s[i-2:i])+96))
                i -= 3
            else:
                new_arr.insert(0, chr(int(s[i])+96))
                i -= 1
        return "".join(new_arr)
```

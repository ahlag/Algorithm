1) Naive Solution

```python
def findComplement(num): 
    return int(''.join(['1' if digit == '0' else '0' for digit in bin(num).replace("0b", "")]),2)

findComplement(5)
```

2) Flip Bit by Bit

```python
Image("complement_5.png",width=700,height=500)
```

```python
def bitwiseComplement(N: int) -> int:
        if N == 0:
            return 1
        todo, bit = N, 1
        while todo:
            # flip current bit
            N = N ^ bit
            # prepare for the next run
            bit = bit << 1
            todo = todo >> 1
        return N
    
bitwiseComplement(5)
```

3) Compute Bit Length and Construct 1-bits Bitmask

```python
Image("complement_7.png",width=700,height=500)
```

```python
from math import log2, floor
def bitwiseComplement(N: int) -> int:
    if N == 0:
        return 1
    # l is a length of N in binary representation
    l = floor(log2(N)) + 1        
    # bitmask has the same length as N and contains only ones 1...1
    bitmask = (1 << l) - 1
    # flip all bits
    return bitmask ^ N

bitwiseComplement(5)
```

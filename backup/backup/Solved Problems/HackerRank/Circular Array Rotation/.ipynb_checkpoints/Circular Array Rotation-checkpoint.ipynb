```python
n,k,q = input().strip().split(' ')
n,k,q = [int(n),int(k),int(q)]
a = [int(a_temp) for a_temp in input().strip().split(' ')]

b = a[:] # copy of the original array so no overwrites occur
for i in range(n):
    b[(i+k)%n] = a[i]
    
for _ in range(q):
    m = int(input())
    print('b[{}] = {}'.format(m,b[m]))
```

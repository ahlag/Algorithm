## O(n^2)
Two iterations

```python
arr = [1,2,3,4,5]

def minimax_sum(arr):
    
    if len(arr) == 0:
        return 0
    
    N = len(arr)
    val = [0] * N
    
    for i in range(N):
        for j in range(N):
            if i == j:
                continue
            else:
                val[i] += arr[j]
    
    print('%d %d' % (min(val), max(val)))

minimax_sum(arr)
```

```python
arr = [1,2,3,4,5]

def minimax_sum(arr):
    
    if len(arr) == 0:
        return 0
    
    N = len(arr)
    max = 0
    min = 10000000000
    sum = 0
    
    for i in range(N):
        sum += arr[i]
        if min > arr[i]:
            min = arr[i]
            
        if max < arr[i]:
            max = arr[i]
    
    print('%d %d' % (sum - max, sum - min))

minimax_sum(arr)
```

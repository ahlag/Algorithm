### some explanations:
#### m is from the equation x1+mv1 = x2+mv2
#### all we need is to find an m that is equal to 0 after modulo. If we can find such m, we are good to go!

```python
def kangaroo(x1, v1, x2, v2):
    
    if v1 > v2:
        n = (x1 - x2) % (v1 - v2)
        if n == 0:
            return 'YES'
        else:
            return 'NO'
    else:
        return 'NO'
    
kangaroo(0,3,4,2)
```

```python
def kangaroo(x1, v1, x2, v2):
    
    if v1 != v2:
        m = (x2-x1)/(v1-v2)
        if m > 0 and m == int(m):
            return ('YES')
        else:
            return ('NO')
    else:
        return ('NO')
    
kangaroo(0,3,4,2)
```

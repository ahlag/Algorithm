```python
#!/bin/python

import math
import os
import random
import re
import sys

# Complete the countSwaps function below.
def countSwaps(a):

    N = len(a)
    cnt = 0

    for i in range(N - 1):
        for j in range(0,N-1-i):
            if a[j] > a[j+1]:
                cnt += 1
                a[j], a[j+1] = a[j+1], a[j] 

    print('Array is sorted in {count} swaps.'.format(count=cnt))
    print('First Element: {first}'.format(first=a[0]))
    print('Last Element: {last}'.format(last=a[-1]))


n = int(input())

a = list(map(int, input().split()))

countSwaps(a)
```

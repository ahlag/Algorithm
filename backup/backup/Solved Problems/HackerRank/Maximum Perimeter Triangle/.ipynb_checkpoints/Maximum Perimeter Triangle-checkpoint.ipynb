```python
sticks = [1,1,1,3,3]

def is_degenerate(a,b,c):
    sides = [a,b,c]
    sides.sort()
    return sum(sides[0:2]) <= sides[-1]

is_degenerate(1,1,1)
```

```python
sticks.sort()
i = len(sticks) - 3

while i >= 0 and (sticks[i] + sticks[i+1] <= sticks[i+2]):
    print(i)
    i -= 1
    
if i > 0:
    print(sticks[i], sticks[i+1], sticks[i+2])
else:
    print(-1)
```

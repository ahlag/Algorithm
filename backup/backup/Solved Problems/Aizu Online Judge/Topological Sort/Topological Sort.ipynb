```python
v, e = map(int, input().split())

adjacency_list = [[] for i in range(v)]

queue = []
deg = [0] * v

for _ in range(e):
    s, t = map(int, input().split())
    adjacency_list[s].append(t)
    deg[t] += 1
    
for i in range(v):
    if deg[i] == 0:
        queue.append(i)
        print(i)
        
while queue:
    vertex = queue.pop(0)
    for v in adjacency_list[vertex]:
        deg[v] -= 1
        if deg[v] == 0:
            queue.append(v)
            print(v)
```

```python
X, N = map(int, input().split())
nums = list(map(int, input().split()))
 
if N == 0:
    print(X)
    exit()
else:
    for i in range(200):
 
        if X-i not in nums:
            print(X-i)
            break
 
        if X+i not in nums:
            print(X+i)
            break
```

```python
from itertools import combinations

n = int(input())
digits = str(n)
m = len(digits)

for i in range(m,0,-1):
    for item in combinations(digits, i):
        if int(''.join(item)) % 3 == 0:
            print(m-len(item))
            exit(0)

## If none, prints(-1)
print(-1)
```

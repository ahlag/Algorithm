```python
# Personal

from itertools import permutations
n = int(input())
v = list(map(int, input().split()))

total = 0
for permutation in list(permutations(v,n)):
    subtotal = permutation[0]
    for i in range(1, len(permutation)):
        subtotal = (subtotal + permutation[i]) / 2
    total = max(total, subtotal)
    
print(total)
```

```python
n = int(input())
v = list(map(int, input().split()))

v.sort()

total = v[0]
for i in range(1, n):
    total = (total + v[i])/ 2
    
print(total)
```

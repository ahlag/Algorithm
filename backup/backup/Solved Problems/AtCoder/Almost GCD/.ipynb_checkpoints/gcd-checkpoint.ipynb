```python
n = int(input())
digits = list(map(int, input().split()))

ans = 0
cnt = 0

for k in range(2, 1001):
    k_cnt = 0
    for digit in digits:
        if digit % k == 0:
            k_cnt += 1
    if k_cnt > cnt:
        cnt = k_cnt
        ans = k
    
print(ans)
```

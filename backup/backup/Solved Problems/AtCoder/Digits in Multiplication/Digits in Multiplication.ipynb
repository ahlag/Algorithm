```python
N = int(input())

def F(a,b):
    return max(len(str(a)), len(str(b)))

ans = float('inf')
for i in range(1, int(N**0.5)+1):
    a = i
    b = N // i
    if N % i == 0:
        ans = min(ans, F(a,b))
    
print(ans)
```

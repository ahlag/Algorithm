```python
from collections import Counter

N = int(input())

def prime_factorization(n):
    table = []
    for i in range(2, int(n ** 0.5) + 1):
        while n % i == 0:
            table.append(i)
            n //= i
    if n > 1:
        table.append(n)
    return table
    
    
T = prime_factorization(N)
C = Counter(T)

cnt = 0
for v in C.values():
    next = 1
    while v >= next:
        v -= 1
        next += 1
        cnt += 1
    
print(cnt)
```

### 訂正: 計算量は O(VN)

```python
# E knapsack2
N, W= map(int, input().split())
V = [[0, 0]]+[list(map(int, input().split())) for _ in range(N)]
dp = [[10**10+1 for _ in range(100001)] for _ in range(N+1)]
dp[0][0]=0
for i in range(1,N+1):
    for j in range(100001):
        # i番目の荷物を選べる
        if j>=V[i][1]:
            dp[i][j] = min(dp[i-1][j],dp[i-1][j-V[i][1]]+V[i][0])
        # i番目の荷物を選ベない
        else:
            dp[i][j] = dp[i-1][j]

dp = dp[-1]
ans = 100000
while dp[ans]>W:
    ans -=1
print(ans)
```

```python
def popularity(votes, N, M):

    total_votes = sum(votes)
    limit = total_votes / (4.0 * M)
    cnt = 0
    for vote in votes:
        if vote >= limit:
            cnt += 1
        if cnt >= M:
            return "Yes"

    return "No"
  
N, M = [int (x) for x in input().split()]
votes = [int (x) for x in input().split()]
print(popularity(votes, N, M))
```

```python
import heapq
import sys

# 
L, P, N = map(int,input().split())
A = list(map(int,input().split()))
B = list(map(int,input().split()))

# 
A += [L]
B += [0]
N += 1

# 
que = []

# ans, pos, tank
ans  = 0
pos  = 0
tank = P

for i in range(N):
    
    # 
    d = A[i] - pos
    
    # 
    while tank - d < 0:
        if len(que) == 0:
            print("-1")
            sys.exit()
        tank += -1 * heapq.heappop(que)
        ans += 1
    
    tank -= d
    pos   = A[i]
    
    heapq.heappush(que, -1 * B[i])

print(ans)
```

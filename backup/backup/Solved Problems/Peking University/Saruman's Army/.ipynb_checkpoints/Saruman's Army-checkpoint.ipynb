```python
N, R = map(int,input().split())
X = list(map(int,input().split()))

i = 0
ans = 0

X.sort()

while i < N:
    
    # s
    s = X[i]
    i += 1
    
    # sR
    while i < N and X[i] <= s + R:
        i += 1
        
    # p
    p = X[i - 1]
    
    # pR
    while i < N and X[i] <= p + R:
        i += 1
    
    ans += 1
        
print(ans)
```

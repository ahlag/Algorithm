```python
# Cosine Rule
import math

def calcAngle(h,m): 
          
        # validate the input 
        if (h < 0 or m < 0 or h > 12 or m > 60): 
            print('Wrong input') 
          
        if (h == 12): 
            h = 0
        if (m == 60): 
            m = 0
          
        # Calculate the angles moved by  
        # hour and minute hands with  
        # reference to 12:00 
        hour_angle = 0.5 * (h * 60 + m) 
        minute_angle = 6 * m 
          
        # Find the difference between two angles 
        angle = abs(hour_angle - minute_angle) 
          
        # Return the smaller angle of two  
        # possible angles 
        angle = min(360 - angle, angle) 
          
        return math.radians(angle)
    
A, B, H, M = list(map(int,input().split()))
angle = calcAngle(H,M)
hypotenuse = math.sqrt(A ** 2 + B ** 2 - 2 * A * B * math.cos(angle))

print('{0:.20f}'.format(hypotenuse))
```

```python
import math

def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    return math.sqrt(dx ** 2 + dy ** 2)

lh, lm, h, m = list(map(int,input().split()))
hx = math.cos(math.pi * 2 * (60 * h + m) / 720) * lh
hy = math.sin(math.pi * 2 * (60 * h + m) / 720) * lh
mx = math.cos(math.pi * 2 * m / 60) * lm
my = math.sin(math.pi * 2 * m / 60) * lm

print('{0:.20f}'.format(distance(hx,hy,mx,my)))
```

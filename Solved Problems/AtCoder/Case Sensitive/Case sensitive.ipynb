```python
# Personal
s = input(); t = input()

if s == t:
    print('same')
else:
    for char1, char2 in zip(s,t):
        res = ord(char1) - ord(char2)

    if res % 32 == 0:
        print('case-insensitive')    
    else:
        print('different')
```

```python
# clean
import sys
input = sys.stdin.readline

def main():
    s = input().strip()
    t = input().strip()
    if s == t:
        print("same")
    elif s.lower() == t.lower():
         print("case-insensitive")
    else:
        print("different")

main()
```

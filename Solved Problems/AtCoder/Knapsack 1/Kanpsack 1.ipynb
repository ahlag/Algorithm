```python
def knapSack(W, wt, val, n): 
    K = [[0 for x in range(W+1)] for x in range(n+1)] 
  
    # Build table K[][] in bottom up manner 
    for i in range(n+1): 
        for w in range(W+1): 
            if i==0 or w==0: 
                K[i][w] = 0
            elif wt[i-1] <= w: 
                K[i][w] = max(val[i-1] + K[i-1][w-wt[i-1]],  K[i-1][w]) 
            else: 
                K[i][w] = K[i-1][w] 
  
    return K[n][W] 
  
n, W = map(int, input().split()) 
val = [] 
wt = [] 

for _ in range(n):
    w, v = map(int, input().split()) 
    val.append(v)
    wt.append(w)

print(knapSack(W, wt, val, n))
```

```python
def knapsack(N, W, wt, val):
    
    K = [[0 for x in range(W+1)] for x in range(N+1)]
    
    for i in range(N+1):
        for w in range(W+1):
            if i == 0 or w == 0:
                K[i][w] == 0
            elif wt[i-1] <= w:
                print('wt[i-1]: {0} w={1}'.format(wt[i-1],w))
                K[i][w] = max(K[i-1][w], K[i-1][w-wt[i-1]] + val[i-1])
            else:
                K[i][w] = K[i-1][w]
        
    return K[N][W]

n, W = map(int, input().split()) 
val = [] 
wt = [] 

for _ in range(n):
    w, v = map(int, input().split()) 
    val.append(v)
    wt.append(w)

print(knapsack(n, W, wt, val))
```

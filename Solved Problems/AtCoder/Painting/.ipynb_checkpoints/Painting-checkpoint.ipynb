```python
H = int(input())
W = int(input())
N = int(input())
 
cnt, paint = 0, 0
while paint < N:
    paint += max(H,W)
    cnt += 1
print(cnt)
```

```python
h = int(input())
w = int(input())
n = int(input())
if n % max(h, w) == 0:
    print(n // max(h, w))
else:
    print(n // max(h, w) + 1)
```

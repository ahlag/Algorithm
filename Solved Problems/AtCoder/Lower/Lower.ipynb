```python
N = int(input())
H = list(map(int, input().split()))

cnt, total = 0, 0

for i in range(1, len(H)):
    if H[i-1] >= H[i]:
        cnt += 1
    else:
        cnt = 0
    total = max(cnt, total)
    
print(total)
```

時間計算量: O(N) \
空間計算量: O(1)

```python
Image("robot_6.png",width=700,height=500)
```

```python
N = int(input())
robots = []
for _ in range(N):
    x, l = map(int, input().split())
    robots.append([x-l, x+l])
 
robots = sorted(robots, key=lambda x: x[1])
now = -float("inf")
ans = 0
# 選べるやつを貪欲に選ぶ
for left, right in robots:
    if left >= now:
        ans += 1
        now = right
print(ans)
```

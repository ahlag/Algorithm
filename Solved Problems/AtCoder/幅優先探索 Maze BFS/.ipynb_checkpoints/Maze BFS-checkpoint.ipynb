```python
R, C = map(int, input().split())
sy, sx = map(int, input().split())
gy, gx = map(int, input().split())

cm = [[0 for j in range(C+1)] for i in range(R+1)]

## Has to start from 1 to avoid index out of range
for i in range(1, R+1):
    s = input()
    for j in range(1, C+1):
        if s[j-1] == '.':
            cm[i][j] =  -1
            
queue = [sy, sx]
cm[sy][sx] = 0
while queue:
    y = queue.pop(0)
    x = queue.pop(0)
    if cm[y][x+1] == -1:
        cm[y][x+1] = 1 + cm[y][x]
        queue.append(y)
        queue.append(x+1)
    if cm[y+1][x] == -1:
        cm[y+1][x] = 1 + cm[y][x]
        queue.append(y+1)
        queue.append(x)
    if cm[y][x-1] == -1:
        cm[y][x-1] = 1 + cm[y][x]
        queue.append(y)
        queue.append(x-1)
    if cm[y-1][x] == -1:
        cm[y-1][x] = 1 + cm[y][x]
        queue.append(y-1)
        queue.append(x)
        
print(cm[gy][gx])
```

```python
S = input()
T = input()

ans = len(T)

for i in range(len(S) - len(T) ):
    diff = 0
    for j in range(len(T)):
        if T[j] != S[i+j]:
            diff += 1
            
    ans = min(ans, diff)

print(ans)
```

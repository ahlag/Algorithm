```python
arr = [1, 100, 1000, 8]

def insertion_sort(arr):
    idx = len(arr) - 2
    target = arr[-1]
    
    while (target < arr[idx]) and (idx >= 0):
        arr[idx+1] = arr[idx]
        print(*arr)
        idx -= 1
      
    print(*arr)
    arr[idx+1] = target
    print(*arr)

insertion_sort(arr)
```

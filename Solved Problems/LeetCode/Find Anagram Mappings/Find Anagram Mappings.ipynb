```python
def anagramMappings(A, B):
    D = {x: i for i, x in enumerate(B)}
    return [D[x] for x in A]
```

```python
A = [12, 28, 46, 32, 50]
B = [50, 12, 32, 46, 28]
anagramMappings(A, B)
```

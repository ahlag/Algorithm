## Personal Solution

```python
class ParkingSystem:

    def __init__(self, big: int, medium: int, small: int):
        self.big = big
        self.cur_big = 0
        self.medium = medium
        self.cur_medium = 0
        self.small = small
        self.cur_small = 0

    def addCar(self, carType: int) -> bool:
        
        if carType == 1:
            if self.cur_big >= self.big:
                return False
            else:
                self.cur_big += 1
                return True
        elif carType == 2:
            if self.cur_medium >= self.medium:
                return False
            else:
                self.cur_medium += 1
                return True
        else:
            if self.cur_small >= self.small:
                return False
            else:
                self.cur_small += 1
                return True
```

## Clean Solution

```python
class ParkingSystem:

    def __init__(self, big: int, medium: int, small: int):
        self.max_slots = [big, medium, small]
        self.slots = [0] * 3 

    def addCar(self, carType: int) -> bool:
        i = carType - 1
        if self.slots[i] == self.max_slots[i]:
            return False

        self.slots[i] += 1
        return True
```

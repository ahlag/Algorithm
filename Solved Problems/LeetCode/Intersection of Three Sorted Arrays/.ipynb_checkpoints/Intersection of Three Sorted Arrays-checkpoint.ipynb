```python
def arrayIntersection(arr1, arr2, arr3):
    sets = [set(arr1), set(arr2), set(arr3)]

    # intersection returns those unique in all list, still need to sort
    return sorted(set.intersection(*sets))
```

```python
A = {1,2,3}
B = {2,3,4}
C = {2}
```

```python
arrayIntersection(A,B,C)
```

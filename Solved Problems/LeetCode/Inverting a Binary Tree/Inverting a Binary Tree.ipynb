```python
## Iterative Approach
def invertTree(root):
    
    if root is None:
        return root
    print(root)
    
    queue = []
    queue.append(root)
    
    while queue:
        node = queue.pop(0)
        
        if node.left is not None:
            queue.append(node.left)
            
        if node.right is not None:
            queue.append(node.right)

        node.left, node.right = node.right, node.left
            
    return root
```

## Personal Solution

```python
class Solution:
    def maximumWealth(self, accounts) -> int:
        
        total = 0
        for account in accounts:
            subtotal = 0
            for i in range(len(account)):
                subtotal += account[i]
            total = max(subtotal, total)
            
        return total
```

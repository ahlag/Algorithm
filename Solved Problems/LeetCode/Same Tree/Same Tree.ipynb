```python
def isSameTree(self, p, q):
    """
    :type p: TreeNode
    :type q: TreeNode
    :rtype: bool
    """    
    # p and q are both None
    if not p and not q:
        return True
    # one of p and q is None
    if not q or not p:
        return False
    if p.val != q.val:
        return False
    return self.isSameTree(p.right, q.right) and \
           self.isSameTree(p.left, q.left)
```

```python
from collections import deque
class Solution:
    def isSameTree(self, p, q):
        """
        :type p: TreeNode
        :type q: TreeNode
        :rtype: bool
        """    
        def check(p, q):
            # if both are None
            if not p and not q:
                return True
            # one of p and q is None
            if not q or not p:
                return False
            if p.val != q.val:
                return False
            return True
        
        deq = deque([(p, q),])
        while deq:
            p, q = deq.popleft()
            if not check(p, q):
                return False
            
            if p:
                deq.append((p.left, q.left))
                deq.append((p.right, q.right))
                    
        return True
```

```python
class Solution(object):
    def isSameTree(self, p, q):
        """
        :type p: TreeNode
        :type q: TreeNode
        :rtype: bool
        """
 
        stack = [[p, q], ]
    
        while stack:
            node1, node2 = stack.pop(0)
            
            if node1 is None and node2 is None:
                continue
                
            if node1 is None or node2 is None or node1.val != node2.val:
                return False
            
            if node1.left is not None or node2.left is not None:
                stack.append([node1.left, node2.left])
            if node1.right is not None or node2.right is not None:
                stack.append([node1.right, node2.right])
                
        return True
```

### Iterative

```python
class Solution(object):
    def isSameTree(self, p, q):
        stack =[]
        stack.append((p,q))

        while stack:
            l,r = stack.pop()

            if not l and not r:
                continue

            if not l or not r or l.val != r.val:
                return False

            stack.append((l.left,r.left))
            stack.append((l.right,r.right))
        return True
```

## Linear Check

```python
class Solution:
    def threeConsecutiveOdds(self, arr) -> bool:
        
        odd_cnt = 0
        for num in arr:
            
            if num % 2 != 0:
                odd_cnt += 1
            else:
                odd_cnt = 0
                
            if odd_cnt == 3:
                return True
            
        return False
```

## Linear Check for three in a row

```python
class Solution:
    def threeConsecutiveOdds(self, arr) -> bool:
        #array length is < 3
        if len(arr) <3:
            return False
        
        for i in range(len(arr) -2):
            if (arr[i] % 2 != 0) and (arr[i+1] % 2 != 0) and (arr[i+2] % 2 != 0):
                return True
        return False
```

```python
class Solution:
    def sortedSquares(self, A):
        
        j = 0
        N = len(A)
        arr = []
        while j < N and A[j] < 0:
            j += 1
        i = j - 1
        
        while i >= 0 and j < N:
            if A[i] ** 2 > A[j] ** 2:
                arr.append(A[j] ** 2)
                j += 1
            else:
                arr.append(A[i] ** 2)
                i -= 1
                
        while i >= 0:
            arr.append(A[i] ** 2)
            i -= 1
        while j < N:
            arr.append(A[j] ** 2)
            j += 1
            
        return arr
```

```python
class Solution:
    def rangeSumBST(self, root: TreeNode, L: int, R: int) -> int:
        
        def rangeSum(node):
            if L <= node.val <= R:
                self.ans += node.val
            if node.left:
                rangeSum(node.left)
            if node.right:
                rangeSum(node.right)
            
        self.ans = 0
        rangeSum(root)
        return self.ans
```

```python
class Solution:
    def rangeSumBST(self, root: TreeNode, L: int, R: int) -> int:
        
        ans = 0
        stack = [root]
        while stack:
            curr = stack.pop()
            if L <= curr.val <= R:
                ans += curr.val
            if curr.right:
                stack.append(curr.right)
            if curr.left:
                stack.append(curr.left)
                
        return ans
```

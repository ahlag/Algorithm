```python
def decompressRLElist(nums):
    """
    :type nums: List[int]
    :rtype: List[int]
    """
    N = len(nums)
    arr = []
    for i in range(0,N,2):
        freq, val = nums[i], nums[i+1]
        arr += [val] * freq
        
    return arr

nums = [1,2,3,4]
decompressRLElist(nums)
```

# Brute Force Method

```python
J = "z"
S = "ZZ"
numJewelsInStones(J,S)
```

```python
J = "aB"
S = "AaBBb"
numJewelsInStones(J,S)
```

Intuition and Algorithm

For each stone, check whether it matches any of the jewels. We can check with a linear scan.

```python
def numJewelsInStones(J, S):
        return sum(s in J for s in S)
```

```python
J = "aB"
S = "AaBBb"
numJewelsInStones(J,S)
```

Complexity Analysis
1) Time Complexity: O(J.length * S.length). \
2) Space Complexity: O(1) additional space complexity in Python. In Java, this can be O(J.length∗S.length)) because of the creation of new arrays.

# Hash Set

```python
def numJewelsInStones(J, S):
        """
        :type J: str
        :type S: str
        :rtype: int
        """
        jewels_collection = set(J)
        cnt = 0
        for stone in S:
            if stone in jewels_collection:
                cnt += 1
                
        return cnt
```

Intuition and Algorithm \
For each stone, check whether it matches any of the jewels. We can check efficiently with a Hash Set.

```python
class Solution(object):
    def numJewelsInStones(self, J, S):
        Jset = set(J)
        return sum(s in Jset for s in S)
```

```python
J = "aB"
S = "AaBBb"
numJewelsInStones(J,S)
```

Complexity Analysis
1) Time Complexity: O(J.length+S.length)). The O(J.length) part comes from creating J. The O(S.length) part comes from searching S. \
2) Space Complexity: O(J.length).

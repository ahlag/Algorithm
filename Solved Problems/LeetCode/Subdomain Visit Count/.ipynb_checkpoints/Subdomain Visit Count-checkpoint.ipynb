```python
class Solution:
    def subdomainVisits(self, cpdomain):
        
        subdomainVisits = {}
        for domains in cpdomains:
            visits, domain = domains.split()
            loop_cnt = domain.count('.')
            for i in range(loop_cnt + 1):
                if domain not in subdomainVisits:
                    subdomainVisits[domain] = int(visits)
                else:
                    subdomainVisits[domain] += int(visits)
                domain = domain.split('.', 1).pop()
        
        output = []
        for k, v in subdomainVisits.items():
            output.append(str(v) + ' ' + str(k))
            
        return output
```

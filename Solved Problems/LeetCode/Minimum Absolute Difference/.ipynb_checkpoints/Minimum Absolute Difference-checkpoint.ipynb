## Sort Array First O(n log n)

```python
def minimumAbsDifference(self, arr):
        arr.sort()
        diff=[]
        for i in range(len(arr)-1):
            diff.append(abs(arr[i+1]-arr[i]))
        d=min(diff)
        ans=[]
        for i in range(len(arr)-1):
            if arr[i+1]-arr[i]==d:
                ans.append([arr[i],arr[i+1]])
        return ans
```

```python
from math import log
class Solution:
    def isPowerOfTwo(self, n: int) -> bool:
        
        if n <= 0:
            return False
        else:
            square_root = log(n)/log(2)
            square_root_rounded = round(square_root)
            diff = abs(square_root - square_root_rounded)
            return True if diff <= 1e-10 else False
```

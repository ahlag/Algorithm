```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
    
        defang = ''
        for i in range(len(address)):
            if address[i] == '.':
                defang += '[.]'
            else:
                defang += address[i]
                
        return defang
```

```python
def defang_ip_address(address):
    return address.replace('.', '[.]')
```

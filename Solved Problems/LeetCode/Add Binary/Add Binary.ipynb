```python
def addBinary(a, b):

    x, y = int(a, 2), int(b, 2)
    
    while y:
        answer = x ^ y
        carry = (x & y) << 1
        x, y = answer, carry
        
    return bin(x)[2:]
```

```python
addBinary("1111", "0010")
```

```python
def balancedStringSplit(s):
    """
    :type s: str
    :rtype: int
    """
    scale = 0
    max_balance = 0
    
    for c in s:

        if(c == 'R'):
            scale += 1
        else:
            scale -= 1
            
        if(scale == 0):
            max_balance += 1
    
    return max_balance

balancedStringSplit("RLRRLLRLRL")
```

```python
def balancedStringSplit(s: str) -> int:
    split = 0
    unbalance = 0
    for i in s:
        unbalance += 1 if i == 'R' else -1    # 'R' = +1, 'L' = -1
        if not unbalance:   # if unbalance == 0
            split += 1
    return split
```

```python
balancedStringSplit("RLRRLLRLRL")
```

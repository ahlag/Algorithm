# Brute Force

```python
def smallerNumbersThanCurrent(nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        max_num = max(nums)
        min_num = min(nums)
        arr = []
        for i in nums:
            cnt = 0
            for j in nums:
                if i > j:
                    cnt += 1
            arr.append(cnt)
            
        return arr
```

## Personal Solution

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums):
        
        n = len(nums)
        sorted_list = sorted(nums)
        
        for i in range(n):
            nums[i] = sorted_list.index(nums[i])
            
        return nums
```

```python
def smallerNumbersThanCurrent(self, nums):
    temp = sorted(nums)
    mapping = {}
    result = []
    for i in range(len(temp)):
        if temp[i] not in mapping:
            mapping[temp[i]] = i
    for i in range(len(nums)):
        result.append(mapping[nums[i]])
    return result
```

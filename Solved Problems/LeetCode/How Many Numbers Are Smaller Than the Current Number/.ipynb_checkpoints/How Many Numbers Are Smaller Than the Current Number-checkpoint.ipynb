# Brute Force

```python
def smallerNumbersThanCurrent(nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        max_num = max(nums)
        min_num = min(nums)
        arr = []
        for i in nums:
            cnt = 0
            for j in nums:
                if i > j:
                    cnt += 1
            arr.append(cnt)
            
        return arr
```

```python
smallerNumbersThanCurrent([1,1,1])
```

```python
smallerNumbersThanCurrent([1,9,1])
```

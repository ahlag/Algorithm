```python
# Iteratively
def reverseList(self, head):
    curr = head
    prev = None
    while curr is not None:
        tempNode = curr.next
        curr.next = prev
        prev = curr
        curr = tempNode
    
    return prev
```

```python
# Recursively    
def reverseList(self, head: ListNode) -> ListNode:
    return self.helper(head, None)
    
    
def helper(self, head, node):
    
    if head is None:
        return node
    
    tempNode = head.next
    head.next = node
    
    return self.helper(tempNode, head)
```

# Bit全探索 
bit全探索とは、bit演算を使って全探索をする方法。bit全探索を使えば、部分集合を全パターン列挙することができる。例えばAtCoderの<a href="https://atcoder.jp/contests/arc061/tasks/arc061_a">「 C - たくさんの数式 / Many Formulas」</a> のような問題で、各文字列中のどこに+を入れるかというパターンを簡単に全探索できる。

### bit全探索とは、2^n通りの全探索！！！

## AND Operator (&)

```python
num = int(input())
for i in range(5):
    print('{}:     {:016b}'.format(num,num))
    print('{}:     {:016b}'.format(i,i))
    print('{} & {}: {:016b}'.format(num,i,num & i))
    print()
```

## Shift Operator (左シフト<</ 右シフト>>)

```python
num = int(input())
for i in range(10):
    print('{} << {}: {:016b} int: {}'.format(num,i,num << i, num << i))
    print()
```

## 注意事項
1 << n  = 2^n

```python
# 入力：１２５
s = ["1", "2", "5"]
n = len(s)
pattern = []

for bit in range(1 << (n - 1)):
    tmp = [False] * (n-1)
    for i in range(n-1):
        if bit & (1 << i):
            tmp[i] = True
    pattern.append(tmp)

print(pattern)
```

```python
# 入力：１２５
s = ["1", "2", "5"]
n = len(s)
pattern = []

for bit in range(2 ** (n - 1)):
    tmp = [False] * (n-1)
    for i in range(n-1):
        if bit & (1 << i):
            tmp[i] = True
    pattern.append(tmp)

print(pattern)
```

```python
s = input()
n = len(s)

ans = 0

for bit in range(1 << (n - 1)):
    f = s[0]

    for i in range(n - 1):
        if bit & (1 << i):
            f += "+"
        f += s[i + 1]

    ans += sum(map(int, f.split("+")))

print(ans)
```

```python
s = input()
n = len(s)

ans = 0

for bit in range(1 << (n - 1)): # 0 1 2 3 (4まで) 1 << (n - 1) = 2 ** (n - 1)
    f = s[0]

    for i in range(n - 1): # 0 1
        if bit & (1 << i): 
            f += "+"
        f += s[i + 1]
    print(f)
    ans += sum(map(int, f.split("+")))

print(ans)
```

## 例題
みかん（100円）りんご（200円）ぶどう（300円）がそれぞれ1つずつ果物屋さんにありました。財布の中には300円ありますが、考え得るすべての買い物パターンを列挙しなさい。

```python
s = input()
n = len(s)
ans = 0
for bit in range(1 << (n-1)):
    f = s[0]
    for i in range(n-1):
        if bit & 1 << i:
            f += '+'
        f += s[i+1]
        
    ans += sum(map(int, f.split('+')))
    
print(ans)
```
